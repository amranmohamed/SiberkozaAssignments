```python
# Firstly i will import necessary libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Dataset Name: Iris Dataset

df = sns.load_dataset('iris')
df.head()
```

```python
# Next, let's visualize the data using a pairplot to see the relationship between different features.
sns.pairplot(df, hue='species')
plt.show()
```

```python
# Now, i'll split the data into training and testing sets.

X = df.drop('species', axis=1)
y = df['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Next, we'll create a decision tree model and fit it to the training data.
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
# Let's evaluate the model's performance on the test data

y_pred = tree.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Finally, let's visualize the decision tree using the plot_tree function from sklearn.tree.

from sklearn.tree import plot_tree
plt.figure(figsize=(25,10))
plot_tree(tree, filled=True, fontsize=10)
plt.show()
```
